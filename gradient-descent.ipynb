#### Title:  Implement Gradient Descent Algorithm to find the local minima of a function. For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.

```python
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return (x + 3) ** 2

def f_derivative(x):
    return 2 * (x + 3)

# Gradient Descent parameters
x = 2                  # Starting point
learning_rate = 0.1    # Step size
epochs = 20            # Number of iterations

# For plotting
x_vals = [x]
y_vals = [f(x)]

# Gradient Descent loop
for i in range(epochs):
    grad = f_derivative(x)
    x = x - learning_rate * grad
    x_vals.append(x)
    y_vals.append(f(x))
    print(f"Step {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

# Final result
print(f"\nLocal minimum occurs at x = {x:.4f}")
print(f"Minimum value of function = {f(x):.4f}")

# Plotting the function and the descent path
x_range = [i for i in range(-6, 3)]
y_range = [f(i) for i in x_range]
plt.plot(x_range, y_range, label="f(x) = (x + 3)^2", color='blue')
plt.scatter(x_vals, y_vals, color='red', label="Gradient Descent Path")
plt.plot(x_vals, y_vals, '--', color='red')
plt.title("Gradient Descent to Find Local Minima")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

### **Conclusion**

The implemented gradient descent algorithm successfully minimizes the function $f(x) = (x + 3)^2$ starting from $x = 2$. Through iterative updates using the function's derivative, the algorithm converges toward the local minimum at $x = -3$. The step-by-step decrease in function value and the plotted descent path visually confirm the effectiveness of gradient descent in finding local minima for convex functions.
